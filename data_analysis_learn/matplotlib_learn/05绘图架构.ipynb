# 绘制折线图plot

```python
#设置图片大小，清晰度
plt.figure(figsize=(10,5),dpi=80)
x = range(2,26,2)
y = [15,14,16,26,85,95,56,32,51,31,12,26]

#设置x，y轴的名字
plt.xlabel("x'slabel")#x轴上的名字
plt.ylabel("y'sabel")#y轴上的名字

#设置x，y作图的范围，有点类似ticks
plt.xlim((0,50))
plt.ylim((-10,2))

#设置x，y轴的刻度
plt.xticks(range(0,30,1))
plt.yticks(range(0,100,5))

#在对应坐标处更换名称
plt.xticks([5,10,15,20,30],["bad","good","great","excelllent","God"])
plt.yticks([0,10,20,50,90],[r'$really\ bad$',r'$b$',r'$c\ \alpha$','d','good'])

plt.plot(x,y)
#plt.savefig("./one.png")#保存图片
plt.show()
```

# 绘制函数：实现figure对象

```python
x = np.linspace(-1,1,50)
y1 = x ** 2
y2 = x * 2
#这个是第一个figure对象,下面的内容都会在第一个figure中显示
plt.figure(num=1)
plt.plot(x,y1)
#这里第二个figure对象
plt.figure(num = 2,figsize = (10,5))
plt.plot(x,y2)
#第三个figure对象
plt.figure(num=3)
l1 = plt.plot(x, y1, label='linear line')
l2 = plt.plot(x, y2, color='red', linewidth=2.0, linestyle='--', label='square line')
plt.show()
```

```python
x = np.linspace(-1,1,50)
y1 = x * 2
y2 = x * x

l1, = plt.plot(x, y1, label='linear line')
l2, = plt.plot(x, y2, color='red', linewidth=2.0, linestyle='--', label='square line')

#简单的设置legend(设置位置)
#位置在右上角
# plt.legend(loc = 'best')
# plt.legend(handles = [l1,l2],labels = ['up','down'],shadow = True,loc = 'upper right')
legend = plt.legend(handles = [l1,l2],labels = ['hu','tang'],loc = 'upper center',shadow = True)
frame = legend.get_frame()
frame.set_facecolor('g') #或者0.9...

# plt.plot(x,y,color='green',linewidth = 3)
plt.xlabel("x'slabel")#x轴上的名字
plt.ylabel("y'sabel")#y轴上的名字
plt.xlim((0,2))
plt.ylim((-2,2))

new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)
#在对应坐标处更换名称
plt.yticks([-2,-1,0,1,2],[r'$really\ bad$',r'$b$',r'$c\ \alpha$','d','good'])

#gca = 'get current axis'
#获取当前的这四个轴
ax = plt.gca()
#设置脊梁(也就是包围在图标四周的默认黑线)
#所以设置脊梁的时候，一共有四个方位
ax.spines['right'].set_color('r')
ax.spines['top'].set_color('r')

#将底部脊梁作为x轴
ax.xaxis.set_ticks_position('bottom')
#ACCEPTS:['top' | 'bottom' | 'both'|'default'|'none']
#设置x轴的位置(设置底的时候依据的是y轴)
ax.spines['bottom'].set_position(('data',0))

#the 1st is in 'outward' |'axes' | 'data'
#axes : precentage of y axis
#data : depend on y data
ax.yaxis.set_ticks_position('left')
# #ACCEPTS:['top' | 'bottom' | 'both'|'default'|'none']
#设置左脊梁(y轴)依据的是x轴的0位置
ax.spines['left'].set_position(('data',0))

plt.show()
```

# 设置四个轴

```python
x = np.linspace(-1,1,50)
# y = x*2
y1 = x * 2
y2 = x * x

l1 = plt.plot(x, y1, label='linear line')
l2 = plt.plot(x, y2, color='red', linewidth=2.0, linestyle='--', label='square line')

#获取当前的这四个轴
ax = plt.gca()
#设置脊梁(也就是包围在图标四周的默认黑线)
#所以设置脊梁的时候，一共有四个方位
ax.spines['right'].set_color('r')
ax.spines['top'].set_color('r')

#将底部脊梁作为x轴
ax.xaxis.set_ticks_position('bottom')
#ACCEPTS:['top' | 'bottom' | 'both'|'default'|'none'|'left'|'right']
#设置x轴的位置(设置底的时候依据的是y轴)
ax.spines['bottom'].set_position(('data',0))

#the 1st is in 'outward' |'axes' | 'data'
#axes : precentage of y axis
#data : depend on y data
ax.yaxis.set_ticks_position('left')
# #ACCEPTS:['top' | 'bottom' | 'both'|'default'|'none'|'left'|'right']
#设置左脊梁(y轴)依据的是x轴的0位置
ax.spines['left'].set_position(('data',0))

#设置坐标轴上的数字风格
# 1. ax.get_xticklabels() 获取得到就是坐标轴上的数字；
# 2. set_bbox() 这个 bbox 就是那坐标轴上的数字的那一小块区域，从结果我们可以很明显的看出
# 来；
# 3. box的facecolor = 'white',edgecolor='none'，第一个参数表示的这个 box 的前面的背景，边上的颜
# 色，第二个使box的边缘颜色；alpha：透明度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor = 'white',edgecolor='green',alpha = 0.7))

plt.show()
```

# 加上箭头标注

```python
x = np.linspace(-3,3,50)
y = 2*x + 1
plt.figure(num = 1,figsize =(8,5))
plt.plot(x,y)
x0 = 1
y0 = 2*x0 + 1

#k--设置虚线
#[x0,x0]设置x范围
#[-5，y0]设置y范围
plt.plot([x0,x0],[-5,y0],'k--',linewidth = 2.5,color="red")#竖线

plt.scatter([x0],[y0],s = 50,color='b')#点

plt.annotate(r'$2x+1 = %s$'% y0,xy = (x0,y0),xycoords = 'data',
            xytext=(+30,-30),textcoords = 'offset points',fontsize = 16
            ,arrowprops = dict(arrowstyle='->',connectionstyle="arc3,rad=.2"))#向量-箭头
            # 1. xy 就是需要进行注释的点的横纵坐标；
            # 2. xycoords = 'data'说明的是要注释点的 xy 的坐标是以横纵坐标轴为基准的；
            # 3. xytext=(+30,-30) 和 textcoords='data'说明了这里的文字是基于标注的点的 x 坐标的偏移 +30 以及标注点 y 坐标 - 30 位置，就是我们要进行注释文字的位置；
            # 4. fontsize = 16 就说明字体的大小
            # 5.arrowprops = dict() 这个是对于这个箭头的描述，arrowstyle='->'这个是箭头的类型。
            # 6.connectionstyle="arc3,rad=.2" 这两个是描述我们的箭头的弧度以及角度的。
plt.show()
```

# zorder控制绘图顺序

```python
x = np.linspace(-3,3,50)
y1 = 0.1*x
y2 = x**2
#zorder控制绘图顺序
#label legend（）的标签名称
plt.plot(x,y1,linewidth = 10,zorder = 1,label = r'$y_1\ =\ 0.1*x$')
plt.plot(x,y2,linewidth = 10,zorder = 2,label = r'$y_2\ =\ x^{2}$')

plt.legend()
plt.show()
```

# 绘制直方图bar

```python
n = 12
X = np.arange(n)
Y1 = (1 - X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1 - X/float(n)) * np.random.uniform(0.5,1.0,n)

#facecolor:表面的颜色;edgecolor:边框的颜色
plt.bar(X,+Y1,facecolor = 'blue',edgecolor = 'green')#向上
plt.bar(X,-Y2,facecolor = 'red',edgecolor = 'white')#向下

#描绘text在图表上
# plt.text(0 + 0.4, 0 + 0.05,"huhu")
for x,y in zip(X,Y1):
    #ha : horizontal alignment
    #va : vertical alignment
    # 前两项是位置，第三项是值
    plt.text(x + 0.01,y+0.05,'%.2f'%y,ha = 'center',va='bottom')
for x,y in zip(X,Y2):
    # ha : horizontal alignment
    # va : vertical alignment
    # 前两项是位置，第三项是值
    plt.text(x+0.01,-y-0.05,'%.2f'%(-y),ha='center',va='top')

plt.xlim(-0.5,n)
plt.yticks([])
plt.ylim(-1.25,1.25)
plt.yticks([])

plt.show()
```

# 3D图

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
#X Y value
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
#hight value
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
"""
============= ================================================
Argument Description
============= ================================================
*X*, *Y*, *Z* Data values as 2D arrays
*rstride* Array row stride (step size), defaults to 10
*cstride* Array column stride (step size), defaults to 10
*color* Color of the surface patches
*cmap* A colormap for the surface patches.
*facecolors* Face colors for the individual patches
*norm* An instance of Normalize to map values to colors
*vmin* Minimum value to map
*vmax* Maximum value to map
*shade* Whether to shade the facecolors
============= ================================================
"""
# I think this is different from plt12_contours
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
"""
========== ================================================
Argument Description
========== ================================================
*X*, *Y*, Data values as numpy.arrays
*Z*
*zdir* The direction to use: x, y or z (default)
*offset* If specified plot a projection of the filled contour
on this position in plane normal to zdir
========== ================================================
"""
ax.set_zlim(-2, 2)
plt.show()
```

# Contours 等高线图

```python
def f(x,y):
    #the height function.
    return (1-x/2 + x**5+y**3) * np.exp(-x **2 -y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
#meshgrid函数用两个坐标轴上的点在平面上画网格。
X,Y = np.meshgrid(x,y)

#use plt.contourf to filling contours
#X Y and value for (X,Y) point
#这里的8就是说明等高线分成多少个部分，如果是0则分成2半
#则8是分成10半
#cmap找对应的颜色，如果高=0就找0对应的颜色值，
plt.contourf(X,Y,f(X,Y),8,alpha = .75,cmap = plt.cm.hot)

#use plt.contour to add contour lines
C = plt.contour(X,Y,f(X,Y),8,colors = 'black',linewidth = .5)

#adding label
plt.clabel(C,inline = True,fontsize = 10)

#ignore ticks
plt.xticks([])
plt.yticks([])

plt.show()
```
