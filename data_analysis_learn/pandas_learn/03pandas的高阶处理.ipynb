##### 缺失值处理

```python
test = pd.read_csv('test.csv')
test
```

```python
# 数据填充
test['单价'].fillna(test['单价'].mean(), inplace=True)
test
```

```python
# 删除缺失行
test.dropna()
```

```python
# 指定替换为 NAN()
test.replace(to_replace='400', value=np.nan)
```

##### 数据离散化

连续的属性值，分成若干离散区间，用不同的符号或整数去表示这些属性

（是否符合正态分布）

```python
test_hist = test['单价']
test_hist.hist(bins=10)
```

```python
# 分区间组
cut = pd.qcut(np.abs(test_hist), 10)
cut
```

```python
# 统计分组次数
cut.value_counts()
```

```python
# 自定义分组
bins = [-1, 1, 2, 3, 9]
p_counts = pd.cut(test_hist, bins)
p_counts.value_counts()
```

```python
# 获得 ‘亚变量’ 矩阵, 分类
m = pd.get_dummies(p_counts, prefix='g')
```
