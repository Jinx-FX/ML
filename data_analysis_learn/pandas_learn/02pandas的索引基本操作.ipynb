# 逻辑运算

```python
# 通过运算符
data['p_change'] > 2
```

```python
data[data['p_change'] > 2]
```

```python
data[(data['p_change']>2) & (data['turnover']>5)]
```

```python
# 查询函数，更加灵活
data.query('p_change>2&turnover>5')
```

```python
# isin 是否存在
data[data['turnover'].isin([4.19, 2.39])]
```

```python
data
```

# 数学运算

```python
open_ = data['open']
close_ = data['close']
```

```python
# add  加法 sub 减法
data['my_price_change'] = close_.sub(open_)
```

```python
data
```
